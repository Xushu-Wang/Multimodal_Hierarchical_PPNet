This takes a image path and displays it

```python
import matplotlib.pyplot as plt
import math
```

```python
paths = [
    f"../output/joint/round-2-.1/images/epoch-10/Diptera/Chironomidae/prototypes/prototype-img-original_with_self_act{i}.png" for i in range(16)
]

w = math.floor(math.sqrt(len(paths)))
h = math.ceil(len(paths) / w)

fig, axs = plt.subplots(h, w, figsize=(20, 20))
for i, path in enumerate(paths):
    img = plt.imread(path)
    ax = axs[i // w, i % w]
    ax.imshow(img)
    ax.axis('off')
fig.suptitle("Diptera>Chironomidae Prototypes", fontsize=30)
fig.tight_layout()
plt.show()
```
