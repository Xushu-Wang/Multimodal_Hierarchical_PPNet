```python
SOURCE_DIR = "/usr/project/xtmp/xw214/full_bioscan_images_2/bioscan/images"
OUTPUT_DIR = "../datasets/full_bioscan_images"
SOURCE_FILE = "../datasets/source_files/metadata_cleaned_permissive.tsv"
```

```python
df = pd.read_csv(SOURCE_FILE, sep="\t")
```

```python
dir_mapping = {}
```

```python
# Iterate over all rows
for index, row in tqdm(df.iterrows()):
    path = os.path.join(row["order"],row["family"], row["genus"], row["species"], row["image_file"])
    dir_mapping[row["image_file"]] = path
```

```python
def copy_file(src, dst):
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy(src, dst)
```

```python
# Walk
for root, dirs, files in os.walk(SOURCE_DIR):
    for file in files:
        src = os.path.join(root, file)
        try:
            dst = os.path.join(OUTPUT_DIR, dir_mapping[file])
            copy_file(src, dst)
        except Exception as e:
            print(f"Error copying {src} to {dst}: {e}")
            continue
```
